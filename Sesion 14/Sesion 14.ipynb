### Agregar Métodos 

```python
class Student(Person):
    def __init__(self, fname, lname, age,language):
        super().__init__(fname, lname, age)
        self.language = language
    
    def __str__(self):        
        return f'{super().__str__()},{self.language}'
```

```python
student1 = Student('Zoro','Roronoa',25,'CHino')
print(student1)
```

### Ejemplo:

Crear una clase se `animal` y luego de la clase `animal` , crear una clase que herede las prodpiedades de `animal`.

```python
class Animal: 
    def __init__(self, name, nfeet, animal_type, skin):
        self.name = name
        self.nfeet = nfeet
        self.animal_type = animal_type
        self.skin = skin
    
    def __str__(self):
        return f'El nombre del animal es: {self.name} el numero de patas {self.nfeet} : y es de tipo {self.animal_type} y esta cubierto de {self.skin}'
```

```python
class Dog(Animal):
    def __init__(self, name, nfeet, animal_type, skin, color):
        super().__init__(name, nfeet, animal_type, skin)
        self.color = color
        
    def __str__(self):
        return f'{super().__str__()},{self.color}'
```

```python
class Cat(Animal):
    def __init__(self, name, nfeet, animal_type, skin, color):
        super().__init__(name, nfeet, animal_type, skin)
        self.color = color
        
    def __str__(self):
        return f'{super().__str__()},{self.color}'
```

```python
class Bird(Animal):
    def __init__(self, name, nfeet, animal_type, skin, color, claws):
        super().__init__( name, nfeet, animal_type, skin)
        self.color = color
        self.claws = claws
        
    def __str__(self):
        return f'{super().__str__()},{self.color},{self.claws}'
```

```python
class Snake(Animal):
    def __init__(self, name, nfeet, animal_type, skin, color, poisson):
        super().__init__(name, nfeet, animal_type, skin)
        self.color = color
        self.poisson = poisson
        
    def __str__(self):
        return f'{super().__str__()},{self.color},{self.poisson}'
```

```python
perro = Dog('chucho',4, 'perro', 'peludo', 'negro')
gato = Cat('miau',4, 'gatos', 'peludo', 'amarillo')
ave = Bird('paquito', 2, 'perico', 'plumas', 'verde', 'chiwquitas')
culebra = Snake('skar', 0, 'cobra', 'piel', 'negro y dorado', 'mucho veneno')
```

```python
print(perro)
print(gato)
print(ave )
print(culebra)
```
